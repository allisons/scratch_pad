```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon, Rectangle
import seaborn as sns
sns.set_style('whitegrid')
```

```python
backlength = 45+1/16
rightlength = 45+.1/16
backdepth = 22+3/16
rightdepth = 20+1/16
```

```python
backmargin = 4
othermargin = 1.5
materialwidth = .75
```

```python
fig, ax = plt.subplots(figsize=(15,15))


ax.set_ylim(-5,50)
ax.set_xlim(-5,50)
ax.set_xticks(np.arange(-5,50))
ax.set_yticks(np.arange(-5,50))
plt.gca().invert_yaxis()
corners_of_desk = np.array([[0,0],[0,backdepth],[backlength-rightdepth,backdepth],[backlength-rightdepth,rightlength],[backlength,rightlength],[backlength,0]])
ax.add_artist(Polygon(corners_of_desk,alpha=0.1))
backdim = backlength-2*othermargin
back = Rectangle([othermargin,backmargin],backdim,materialwidth,facecolor='purple',edgecolor="Black",alpha=.5)
frontdim = backdim
front = Rectangle([othermargin,backdepth-othermargin],frontdim,materialwidth,facecolor='blue',edgecolor="Black",alpha=.5)
leftdim = backdepth-(materialwidth*3+backmargin)
left = Rectangle([othermargin,backmargin+materialwidth],materialwidth,leftdim,facecolor='blue',edgecolor="Black",alpha=.5)
rightdim = leftdim
right = Rectangle([backlength-(othermargin+materialwidth),backmargin+materialwidth],materialwidth,rightdim,facecolor='blue',edgecolor="Black",alpha=.5)
leftfrontdim=rightlength-backdepth-materialwidth
leftfront = Rectangle([backlength-rightdepth+othermargin,backdepth-othermargin+materialwidth],materialwidth,leftfrontdim,facecolor='blue',edgecolor="Black",alpha=.5)
rightfrontdim = leftfrontdim
rightfront = Rectangle([backlength-(othermargin+materialwidth),backdepth-othermargin+materialwidth],materialwidth,rightfrontdim,facecolor='blue',edgecolor="Black",alpha=.5)
frontrightdim = rightdepth-othermargin*2
frontright = Rectangle([backlength-rightdepth+othermargin,rightlength-othermargin],frontrightdim,materialwidth,facecolor='blue',edgecolor="Black",alpha=.5)
ax.add_artist(back)
ax.add_artist(front)
ax.add_artist(left)
ax.add_artist(right)
ax.add_artist(leftfront)
ax.add_artist(rightfront)
ax.add_artist(frontright)
hor = [((backlength/2,backmargin),backdim),
              ((backlength/2,backdepth-othermargin-materialwidth),backdim),
              ((backlength-rightdepth/2,rightlength-othermargin),frontrightdim)]
for (x,y),length in hor:
    ax.annotate(length,xy=(x,y))
vert = [((othermargin-materialwidth,backdepth/2),leftdim),
       ((backlength-othermargin-materialwidth*2,backdepth/2),leftdim),
       ((backlength-rightdepth,rightlength-leftfrontdim/2),leftfrontdim),
        ((backlength-othermargin,rightlength-leftfrontdim/2),rightfrontdim)]
for (x,y),length in vert:
    ax.annotate(length,xy=(x,y),rotation=90)
```
